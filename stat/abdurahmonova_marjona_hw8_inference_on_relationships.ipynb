1. У случайной выборки людей разного возраста спросили надевают ли они очки при чтении. Результаты приведены в таблице ниже. Имеется ли взаимосвязь между возрастом и ношением очков?
   
  |  | надевает очки | не надевает |
  |--|---------------|-------------|
  | возраст $\leq$ 20 | 15 | 56 |  
  | 20 < возраст $\leq$ 40 | 19 | 37 |  
  | 40 < возраст $\leq$ 60 | 34 | 35 |  
  | возраст > 60 | 51 | 2

```python
import scipy.stats as stats
import numpy as np
```

```python
observed=np.array([
    [15, 56],
    [19, 37],
    [34, 35],
    [51, 2]
])

chi2, p_value, dof, expected=stats.chi2_contingency(observed)

print(f"Статистика хи-квадрат: {chi2:.4f}")
print(f"P-значение: {p_value:.4f}")

alpha=0.05
if p_value<alpha:
    print("Есть статистически значимая взаимосвязь между возрастом и ношением очков.")
else:
    print("Нет достаточных оснований утверждать, что возраст связан с ношением очков.")
```

2. Случайная выборка студентов была разделена на 3 группы и каждая из них готовилась к сдаче экзамена по одной из 3 методик (A, B, C). В таблице ниже приведены оценки по экзамену студентов каждой группы. Определите имеются ли различия между группами.
 
| Group 1 | Group 2 | Group 3 |
|---------|---------|---------|
| 85      | 91      | 79      |
| 86      | 92      | 78      |
| 88      | 93      | 88      |
| 75      | 85      | 94      |
| 78      | 87      | 92      |
| 94      | 84      | 85      |
| 98      | 82      | 83      |
| 79      | 88      | 85      |
| 71      | 95      | 82      |
| 80      | 96      | 81      |

```python
group1=[85, 86, 88, 75, 78, 94, 98, 79, 71, 80]
group2=[91, 92, 93, 85, 87, 84, 82, 88, 95, 96]
group3=[79, 78, 88, 94, 92, 85, 83, 85, 82, 81]

f_statistic, p_value=stats.f_oneway(group1, group2, group3)

print(f"F-статистика: {f_statistic:.4f}")
print(f"P-значение: {p_value:.4f}")

alpha=0.05
if p_value<alpha:
    print("Есть статистически значимые различия между группами.")
else:
    print("Нет статистически значимых различий между группами.")
```

```python
before_training=np.array([55.88, 50.80, 48.26, 60.96, 63.50, 63.50, 71.12, 55.88, 76.20, 68.58, 
                           60.96, 45.72, 40.64, 48.26, 48.26, 71.12, 60.96, 63.50, 63.50, 58.42])

after_training=np.array([60.96, 55.88, 48.26, 55.88, 71.12, 66.04, 71.12, 60.96, 76.20, 73.66, 
                           63.50, 50.80, 43.18, 45.72, 45.72, 71.12, 66.04, 68.58, 68.58, 60.96])

t_statistic, p_value=stats.ttest_rel(before_training, after_training)

print(f"t-статистика: {t_statistic:.4f}")
print(f"P-значение: {p_value:.4f}")

alpha=0.05
if p_value<alpha:
    print("Программа тренировок улучшила высоту вертикального прыжка.")
else:
    print("Программа тренировок не оказала значительного эффекта на высоту вертикального прыжка.")
```
